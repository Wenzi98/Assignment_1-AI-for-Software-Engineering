```python
# EcoWealth Advisor - Sustainable Cryptocurrency Investment Assistant
crypto_db = {
    "Bitcoin": {
        "price_trend": "rising",
        "market_cap": "high",
        "energy_use": "high",
        "sustainability_score": 3/10
    },
    "Ethereum": {
        "price_trend": "stable",
        "market_cap": "high",
        "energy_use": "medium",
        "sustainability_score": 6/10
    },
    "Cardano": {
        "price_trend": "rising",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 8/10
    },
    "Solana": {
        "price_trend": "rising",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 9/10
    },
    "Algorand": {
        "price_trend": "stable",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 9/10
    },
    "Polkadot": {
        "price_trend": "falling",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 7/10
    }
}

def eco_wealth_advisor():
    print("""
    üåøüíπ Welcome to EcoWealth Advisor! üíπüåø
    Your AI-powered financial sidekick for sustainable crypto investments.
    
    I analyze cryptocurrencies based on:
    - Profitability üìà (price trends, market cap)
    - Sustainability ‚ôªÔ∏è (energy efficiency, project viability)
    
    Try asking me:
    - "What's trending?"
    - "Show me eco-friendly options"
    - "Recommend sustainable coins with growth potential"
    - "Tell me about Solana"
    - "Exit" to end our session
    
    ‚ö†Ô∏è Important: Cryptocurrency investments carry significant risk. 
    This is not financial advice‚Äîalways conduct your own research!
    """)
    
    while True:
        user_query = input("\nYou: ").strip().lower()
        
        # Exit condition
        if user_query in ['exit', 'quit', 'bye']:
            print("EcoWealth: Remember to diversify and invest responsibly! Good luck! üåü")
            break
        
        # Greeting response
        elif any(word in user_query for word in ['hi', 'hello', 'hey']):
            print("EcoWealth: Hello! Ready to explore sustainable crypto investments?")
        
        # Help menu
        elif 'help' in user_query or 'what can you do' in user_query:
            print("EcoWealth: I can help you:\n"
                  "- Identify trending cryptocurrencies üìà\n"
                  "- Find eco-friendly coins ‚ôªÔ∏è\n"
                  "- Recommend coins balancing profit & sustainability\n"
                  "- Provide details about specific cryptocurrencies\n"
                  "Try asking: 'Recommend green coins with growth potential'")
        
        # Sustainability recommendation
        elif any(word in user_query for word in ['sustainab', 'eco', 'green', 'environment']):
            sustainable_coins = [(coin, data) for coin, data in crypto_db.items() 
                                if data["energy_use"] == "low" and data["sustainability_score"] >= 7/10]
            
            if sustainable_coins:
                sustainable_coins.sort(key=lambda x: x[1]["sustainability_score"], reverse=True)
                response = "Top sustainable coins:\n"
                for coin, data in sustainable_coins[:3]:
                    response += f"- {coin}: ‚ôªÔ∏è Score {data['sustainability_score']*10}/10, Energy {data['energy_use'].capitalize()}\n"
                print(response)
            else:
                print("EcoWealth: No highly sustainable coins meet current criteria. Consider Cardano or Algorand.")
        
        # Profitability recommendation
        elif any(word in user_query for word in ['trend', 'profit', 'grow', 'rising']):
            rising_coins = [coin for coin, data in crypto_db.items() 
                           if data["price_trend"] == "rising"]
            
            if rising_coins:
                profitable_coins = sorted(
                    [(coin, crypto_db[coin]) for coin in rising_coins],
                    key=lambda x: (x[1]["market_cap"] == "high", x[1]["sustainability_score"]),
                    reverse=True
                )
                response = "Currently trending coins:\n"
                for coin, data in profitable_coins[:3]:
                    response += f"- {coin}: üìà Rising, Cap: {data['market_cap'].capitalize()}\n"
                print(response)
            else:
                print("EcoWealth: Market appears stable. Consider sustainable options for long-term holds.")
        
        # Combined recommendation
        elif any(word in user_query for word in ['both', 'balance', 'recommend', 'advise']):
            ideal_coins = [(coin, data) for coin, data in crypto_db.items() 
                          if data["price_trend"] == "rising" 
                          and data["sustainability_score"] >= 6/10]
            
            if ideal_coins:
                ideal_coins.sort(key=lambda x: x[1]["sustainability_score"], reverse=True)
                best_coin = ideal_coins[0][0]
                details = crypto_db[best_coin]
                print(f"EcoWealth: Based on growth and sustainability, consider {best_coin}:\n"
                      f"- Trend: {details['price_trend'].capitalize()}\n"
                      f"- Sustainability: ‚ôªÔ∏è {details['sustainability_score']*10}/10\n"
                      f"- Energy: {details['energy_use'].capitalize()}\n"
                      f"- Market Cap: {details['market_cap'].capitalize()}")
            else:
                print("EcoWealth: Current market conditions make balanced recommendations challenging. "
                      "Cardano remains a strong sustainable option.")
        
        # Coin-specific inquiry
        elif any(word in user_query for word in ['about', 'tell me', 'what']):
            found = False
            for coin in crypto_db:
                if coin.lower() in user_query:
                    details = crypto_db[coin]
                    print(f"EcoWealth: {coin} Analysis:\n"
                          f"- Price Trend: {details['price_trend'].capitalize()}\n"
                          f"- Market Cap: {details['market_cap'].capitalize()}\n"
                          f"- Energy Use: {details['energy_use'].capitalize()}\n"
                          f"- Sustainability: ‚ôªÔ∏è {details['sustainability_score']*10}/10\n"
                          f"Verdict: {'‚úÖ Good balance' if details['sustainability_score'] >= 6 and details['price_trend'] == 'rising' else '‚ö†Ô∏è Consider alternatives'}")
                    found = True
                    break
            if not found:
                print("EcoWealth: Sorry, I don't have data on that cryptocurrency. Try Bitcoin, Ethereum, or Cardano.")
        
        # Fallback response
        else:
            print("EcoWealth: I'm not sure I understand. Try asking about:\n"
                  "- Trending cryptocurrencies\n"
                  "- Eco-friendly coins\n"
                  "- Specific coins like Cardano\n"
                  "Or type 'help' for options")

# Start the advisor
if __name__ == "__main__":
    eco_wealth_advisor()
```
